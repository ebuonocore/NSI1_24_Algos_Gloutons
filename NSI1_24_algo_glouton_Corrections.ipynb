## Trier la table

```python
def rapport_valeur_poids(dico_objet):
    """ Prend en paramètre un dictionnaire dico_objet 
        Renvoie le rapport de la valeur divisée par le poids de l'objet.
    """
    return dico_objet['valeur']/dico_objet['poids']
```

```python
print( rapport_valeur_poids({'nom' : 'objet D', 'poids' : 25, 'valeur' : 750}))
# On  s'attend à obtenir 750 / 25 soit 30 unité de valeur par unité de coût
print( rapport_valeur_poids({'nom' : 'objet G', 'poids' : 2,  'valeur' : 1400}))
# On  s'attend à obtenir 1400 / 2 soit 700 unité de valeur par unité de coût
# On en déduit qu'il est préféreble de sélectionner le second objet a priori
```

```python
def donner_poids(dico_objet):
    """ Prend en paramètre un dictionnaire dico_objet
        Renvoie le poids de l'objet
    """
    return dico_objet['poids']
```

**Question:**   
    
    
Créer une fonction `creer_table_triee` qui prend en paramètre une table d'objets `table_objets` (table similaire aux deux tables `table_1` et `table_2` ci-dessus) et renvoie une copie triée en deux étapes de cette table :  

- triee selon le poids des objets décroissant  


- puis triee selon le rapport valeur/poids décroissant  
    
    
On utilisera pour cela les deux fonctions clef de tri `donner_poids` et `rapport_valeur_sur_poids`.
    


**Si besoin, retourner voir son cours et ses exercices sur le traitement de donnée en tables.**


```python
def creer_table_triee(table_objets):
    """ Prend en paramètre une table d'objets table_objets
        Trie selon le poids des objets décroissant
        Puis trie selon le rapport valeur/poids décroissant
    """
    table_apres_premier_tri = sorted(table_objets, key = donner_poids, reverse = True)
    table_apres_second_tri = sorted(table_apres_premier_tri, key = rapport_valeur_poids, reverse = True)
    """ Le premier tri prépare la liste pour que le second tri place l'objet le moins lourd
        devant  en cas d'égalité du rapport valeur / poids.
        On recréé la liste des objets dans l'ordre croissant du rapport valeur/coût.
    """
    return table_apres_second_tri
```

```python
def affiche_resultat_glouton(table):
    t = creer_table_triee(table) 
    # t correspond à la liste des objets de table triée selon le rapport valeur / coût
    # On affiche chaque élément de t suivi du rapport valeur / coût. 
    somme_poids = 0 
    for element in t:
        somme_poids += donner_poids(element)
        print(element,'rapport valeur / coût = ', rapport_valeur_poids(element))
    # On vérifie bien que les rapport valeur / coût sont décroissants
    print ("Poids total des objets sélectionnés", somme_poids)
    # Pour finir, on affiche le poids total des objets sélectionnés
    
affiche_resultat_glouton(table_1)
```

## Parcourir la table triée et sélectionner les objets

**Question:**  

Compléter la fonction `selectionner` qui prend en paramètre une table `table_objets` (table similaire aux deux tables `table_1` et `table_2` ci-dessus) ainsi qu'un poids maximal `poids_max` et retourne dans une table `table_selection` la sélection d'objets obtenue selon l'algorithme glouton.  
    
 
On rappelle que pour ajouter un élément `elt` dans une liste `L` on peut utiliser l'instruction `L.append(elt)`.

```python
def selectionner( table_objets, poids_max):
    """ Prend en paramètre une table table_objets, ainsi qu'un poids maximal poids_max
        Renvoie dans une table table_selection la sélection d'objets obtenue selon l'algorithme glouton
    """
    table_triee = creer_table_triee(table_objets)
    # La liste est triée selon le rapport valeur / coût décroissant
    poids_total = 0
    table_selection = []
    # On initialise les variables. Le sac est vide: Aucun élément, poids nul.
    for objet in table_triee:
        # Pour chaque objet de la liste triée ...
        if poids_total + donner_poids(objet) <= poids_max :
            # Si l'objet arrive à rentrer
            # Alorson ajoute son 'poids' à la somme des poids des objets dans le sac
            poids_total = poids_total + donner_poids(objet)
            # et on le rajoute à la liste des objets entrés 
            table_selection.append(objet)
    return table_selection
```

```python
# On répond à l'énoncé du problème 1
print("\n Exemple pour une sélection d'objet de la table_1 avec un sac de poids_max = 40")
affiche_resultat_glouton( selectionner( table_1, 40) )

# On répond à l'énoncé du problème 1
print("\n Exemple pour une sélection d'objet de la table_2 avec un sac de poids_max = 8")
affiche_resultat_glouton( selectionner( table_2, 8)  )
```
